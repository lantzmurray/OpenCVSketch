```python
### Let's import our libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
### Let's load up an image of our family dog, Axel
img = cv2.imread('/content/CV Sketch Project/MrAxel.jpeg')
plt.imshow(img)
```

```python
###The color is all messed up.  Axel and the floor should be a shade of brown. 
###His tie should be blue. Lets change the colors from BGR to RGB 

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
###Let's convert the image to grayscale 

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
### Lets add some Gaussian Blur to the image
#This will make it look smooth, or blurry. 


blur = cv2.GaussianBlur(gray,(5,5),2)
plt.imshow(blur, cmap ='gray')
```

```python
sobelx=cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=5)
sobely=cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=5)

gradmag_sq = np.square(sobelx)+np.square(sobely)
gradmag = np.sqrt(gradmag_sq)

plt.imshow(gradmag,cmap='gray')
```

```python
gradmag_inv = 255-gradmag
plt.imshow(gradmag_inv,cmap='gray')
```

```python
thresh_value , thresh_img = cv2.threshold(gradmag_inv,6,255,cv2.THRESH_BINARY)
plt.imshow(thresh_img,cmap='gray')
```

```python
pencilsketchgrey, pencilsketchcolor =cv2.pencilSketch(img_rgb,sigma_s=60,sigma_r=.09,shade_factor=.05)
plt.imshow(pencilsketchgrey,cmap='gray')
```

```python
plt.imshow(pencilsketchcolor)
```

```python
waterpainting = cv2.stylization(img_rgb,sigma_s=3,sigma_r=3)
plt.imshow(waterpainting)
```
